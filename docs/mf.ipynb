# Membership Functions

A membership function (*MF*) assigns a membership value between 0 and 1 to each point in the input space, indicating the degree to which that point belongs to a fuzzy set. This input space is called the universe of discourse (*UoD*). A degree of membership of 0 means "not a member" and 1 means "fully a member" of the fuzzy set.

```python
import altair as alt

from mostly.membership_functions import MFTrapezoidal, MFTriangular
from mostly.plotting.altair.plot_fuzzy_set import plot_fuzzy_set

alt.renderers.set_embed_options(actions=False)
```

## Piecewise Linear Functions

### Z-shaped

```python
plot_fuzzy_set(
    mf=MFTrapezoidal(a=2, b=2, c=2, d=8),
    term="good",
    uod=(0, 10),
)
```

Use same logic to model fully linear memberships

```python
plot_fuzzy_set(
    mf=MFTrapezoidal(a=0, b=0, c=0, d=10),
    term="good",
    uod=(0, 10),
)
```

### S-Shaped

```python
plot_fuzzy_set(
    mf=MFTrapezoidal(a=2, b=8, c=8, d=8),
    term="good",
    uod=(0, 10),
)
```

### Triangular

```python
plot_fuzzy_set(
    mf=MFTriangular(a=0, b=5, c=10),
    term="good",
    uod=(0, 10),
)
```

Use `MFTriangular` or `MFTrapezoidal` to model a triangular.

```python
plot_fuzzy_set(
    mf=MFTrapezoidal(a=0, b=5, c=5, d=10),
    term="good",
    uod=(0, 10),
)
```

### Trapezoidal

```python
plot_fuzzy_set(
    mf=MFTrapezoidal(a=0, b=3, c=5, d=10),
    term="good",
    uod=(0, 10),
)
```
